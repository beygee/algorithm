```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as npr
```

```python
plt.plot([3,1,5,2])
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)
ax1 = plt.gca()
ax2 = plt.gca().twinx()

ax1.plot(x, np.sin(x))
ax2.plot(x, 10 * np.cos(2 * x), color = 'y')

plt.show()
```

```python
x1 = np.linspace(0, 2 * np.pi, 50)
y1 = np.sin(x1)

x2 = np.linspace(0, 2 * np.pi, 50)
y2 = np.tan(x2)

plt.subplot(1, 2, 1)
plt.plot(x1, y1)

plt.subplot(1, 2, 2)
plt.plot(x2, y2)

plt.show()
```

```python
y1 = npr.normal(size = 50).cumsum()
y2 = npr.normal(size = 100).cumsum()
y3 = npr.normal(size = 50).cumsum()
y4 = npr.normal(size = 50).cumsum()

plt.subplot(2,2,1)
plt.plot(y1)

plt.subplot(2,2,2)
plt.plot(y2)

plt.subplot(2,2,3)
plt.plot(y3)

plt.subplot(2,2,4)
plt.plot(y4)

plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)

plt.plot(np.sin(x), label="sin")
plt.plot(np.cos(x), label="cos")

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, np.sin(x))
plt.title("Title: sin(x)", fontsize=15)
plt.xlabel('X: radians')
plt.ylabel('Y: amplitude')

plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, np.sin(x))

plt.grid()
```

```python
y1 = npr.normal(size = 50).cumsum()

plt.subplot(1,2,1)
plt.plot(y1)
plt.title("Y1")

plt.subplot(1,2,2)
plt.plot(y1)
plt.xlim(0, 100)
plt.ylim(-10,10)
plt.title("Y2")

plt.show()
```
